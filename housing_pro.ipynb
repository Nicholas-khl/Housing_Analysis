## Data Manipulation

```python
housing_s_copy = housing_s
housing_s
```

```python
housing_s_copy.info()
housing_s_copy.columns    ### Dtype = object creates diffculties for plotting; should contain int dtype
```

```python
for index, row in housing_s_copy.iterrows():
    housing_s_copy.at[index, "Total units"] = row["Total units"].replace(',', '')
    housing_s_copy.at[index, "Single-detached"] = row["Single-detached"].replace(',', '')
    housing_s_copy.at[index, "Multiples"] = row["Multiples"].replace(',', '')
    housing_s_copy.at[index, "Semi-detached"] = row["Semi-detached"].replace(',', '')
    housing_s_copy.at[index, "Row"] = row["Row"].replace(',', '')
    housing_s_copy.at[index, "Apartment and other unit type"] = row["Apartment and other unit type"].replace(',', '')

for index, row in housing_uc.iterrows():
    housing_uc.at[index, "Total units"] = row["Total units"].replace(',', '')
    housing_uc.at[index, "Single-detached"] = row["Single-detached"].replace(',', '')
    housing_uc.at[index, "Multiples"] = row["Multiples"].replace(',', '')
    housing_uc.at[index, "Semi-detached"] = row["Semi-detached"].replace(',', '')
    housing_uc.at[index, "Row"] = row["Row"].replace(',', '')
    housing_uc.at[index, "Apartment and other unit type"] = row["Apartment and other unit type"].replace(',', '')

for index, row in housing_c.iterrows():
    housing_c.at[index, "Total units"] = row["Total units"].replace(',', '')
    housing_c.at[index, "Single-detached"] = row["Single-detached"].replace(',', '')
    housing_c.at[index, "Multiples"] = row["Multiples"].replace(',', '')
    housing_c.at[index, "Semi-detached"] = row["Semi-detached"].replace(',', '')
    housing_c.at[index, "Row"] = row["Row"].replace(',', '')
    housing_c.at[index, "Apartment and other unit type"] = row["Apartment and other unit type"].replace(',', '')
```

```python
annual_stat = []

for year in range(2014, 2023):
    annual_total = housing_s_copy[housing_s_copy["Quarter"].str.contains(str(year))]["Total units"].sum()   ###find year in quarter and use panda fuction to sum it up & hard to do with 2 for-loops
    annual_total_4 = housing_c[housing_c["Quarter"].str.contains(str(year))]["Total units"].sum()
    annual_total_2 = population[housing_s_copy["Quarter"].str.contains(str(year))]["difference"].sum()
    annual_total_3 = round(annual_total_2 / average_household, 1)
    #print(f'The annual number of housing starts in {year} is {annual_total}, the growth of population is {annual_total_2}, and the amount of housing needed is {annual_total_3}')

    annual_stat.append([year, annual_total, annual_total_4, annual_total_2, annual_total_3])

annual_df = pd.DataFrame(annual_stat, columns=["Year", "Housing_starts", "Housing_finished", "Population_growth", "Amount_of_housing_needed"])
annual_df
```
